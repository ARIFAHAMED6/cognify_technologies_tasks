```python

'''
Task 3: Create a console application
for basic CRUD operations on a list
of tasks.
Objective: Implement Create, Read, Update, and
Delete operations using arrays or lists for data
storage.
Steps:
1.Define a Task class with necessary attributes.
2.Implement functionality to create a new task.
3.Develop a method to read and display tasks.
4.Allow users to update task details.
5.Provide an option to delete tasks.
6.Test the application with various scenarios.
'''
```

```python
class Task:
    def __init__(self, task_id, title, description):
        self.id = task_id
        self.title = title
        self.description = description

tasks = []
task_id_counter = 1

while True:
    print("\n TASK MANAGER MENU")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Update Task")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice =='1':
        title = input("Enter task title: ")
        description = input("Enter task description: ")
        task = Task(task_id_counter, title, description)
        tasks.append(task)
        print(f" Task {task_id_counter} added.")
        task_id_counter += 1

    elif choice =='2':
        if not tasks:
            print("No tasks available.")
        else:
            print("\n Task List:")
            for task in tasks:
                print(f"ID: {task.id} | Title: {task.title} | Description: {task.description}")

    elif choice =='3':
        task_id = int(input("Enter task ID to update: "))
        found = False
        for task in tasks:
            if task.id == task_id:
                task.title = input("Enter new title: ")
                task.description = input("Enter new description: ")
                print("Task updated.")
                found = True
                break
        if not found:
            print("Task ID not found.")

    elif choice =='4':
        task_id = int(input("Enter task ID to delete: "))
        found = False
        for task in tasks:
            if task.id == task_id:
                tasks.remove(task)
                print("Task deleted.")
                found = True
                break
        if not found:
            print("Task ID not found.")

    elif choice =='5':
        print(" Exiting Task Manager. Thank you!")
        break

    else:
        print(" Invalid input. Please enter a number from 1 to 5.")
```
